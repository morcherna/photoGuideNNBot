```python
def get_class_names():
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM landmarks")
    classes = [row[0] for row in cursor.fetchall()]
    conn.close()
    return classes
```

```python
def predict_image_class(image: Image.Image) -> str:
    class_names = get_class_names()

    model = models.efficientnet_b0(weights=None)
    model.classifier[1] = torch.nn.Linear(model.classifier[1].in_features, len(class_names))
    model.load_state_dict(torch.load(model_path))
    model.eval()

    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ])
    input_tensor = transform(image).unsqueeze(0)

    with torch.no_grad():
        output = model(input_tensor)
        _, predicted = torch.max(output, 1)

    return class_names[predicted.item()]
```

```python
def get_info_from_db(name):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT description, photo_url FROM landmarks WHERE name=?", (name,))
    row = cursor.fetchone()
    conn.close()
    return {"описание": row[0], "фото": row[1]}
```

```python
def save_image(image, class_name, filename, save_dir=save_dir):
    class_dir = os.path.join(save_dir, class_name.strip())
    os.makedirs(class_dir, exist_ok=True)
    image.save(os.path.join(class_dir, filename))
```

```python
def generate_summary(description):
    prompt = f"Сделай краткое, но информативное описание 1-2 предложения не более 30 слов следующего текста: {description}"
    try:
        response = gpt_client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}],
            web_search=False
        )
        return response.choices[0].message.content
    except Exception as e:
        print("Ошибка при генерации краткого описания:", e)
        return "Произошла ошибка при создании краткого описания."
```

```python
graph = ox.graph_from_place("Нижний Новгород, Россия", network_type="walk")
```

```python
def get_filtered_places(user_location, selected_categories, radius_km, max_count):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    placeholder = ",".join("?" for _ in selected_categories)
    query = f"""
    SELECT l.id, l.name, l.latitude, l.longitude
    FROM landmarks l
    JOIN landmark_categories lc ON l.id = lc.landmark_id
    JOIN categories c ON lc.category_id = c.id
    WHERE c.name IN ({placeholder})
    GROUP BY l.id
    """
    cursor.execute(query, selected_categories)
    results = cursor.fetchall()
    conn.close()

    filtered = []
    for row in results:
        landmark_id, name, lat, lon = row
        distance = geodesic(user_location, (lat, lon)).km
        if distance <= radius_km:
            filtered.append({
                "id": landmark_id,
                "name": name,
                "location": (lat, lon),
                "distance": distance
            })

    filtered.sort(key=lambda x: x["distance"])
    return filtered[:max_count]
```

```python
def html_to_png_with_selenium(html_path, output_path='route.png'):
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    driver = webdriver.Chrome(options=options)
    file_url = 'file://' + os.path.abspath(html_path)
    driver.get(file_url)
    time.sleep(2)

    driver.set_window_size(1024, 768)
    driver.save_screenshot(output_path)
    driver.quit()
```

```python
def plot_route_to_places(user_location, places, message_chat_id=None):
    orig_node = ox.distance.nearest_nodes(graph, user_location[1], user_location[0])
    map = folium.Map(location=user_location, zoom_start=13)
    folium.Marker(location=user_location, popup="Ваше местоположение", icon=folium.Icon(color='blue')).add_to(map)

    min_lat, min_lon = user_location
    max_lat, max_lon = user_location

    for place in places:
        name = place['name']
        place_location = place['location']

        dest_node = ox.distance.nearest_nodes(graph, place_location[1], place_location[0])
        route = nx.shortest_path(graph, orig_node, dest_node, weight='length')

        route_coords = [(graph.nodes[n]['y'], graph.nodes[n]['x']) for n in route]
        folium.PolyLine(route_coords, color='blue', weight=5).add_to(map)

        folium.Marker(location=place_location, popup=name, icon=folium.Icon(color='red')).add_to(map)
        orig_node = dest_node

        min_lat = min(min_lat, place_location[0])
        min_lon = min(min_lon, place_location[1])
        max_lat = max(max_lat, place_location[0])
        max_lon = max(max_lon, place_location[1])

    map.fit_bounds([[min_lat, min_lon], [max_lat, max_lon]])

    map_html_path = 'route_map.html'
    map.save(map_html_path)

    map_png_path = 'route.png'
    html_to_png_with_selenium(map_html_path, map_png_path)

    if message_chat_id:
        user_state[str(message_chat_id)] = user_state.get(str(message_chat_id), {})
        user_state[str(message_chat_id)]['route_places'] = [place['name'] for place in places]

    return map_html_path, map_png_path
```

```python
import csv

def save_rating(user_id, data):
    with open("ratings.csv", "a", newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow([user_id, data['recognition'], data['routes'], data['speech'], data['summary'], data['total']])
```

```python
@bot.message_handler(commands=['rate'])
def start_survey(message):
    chat_id = message.chat.id
    user_state[chat_id] = {'step': 1, 'data': {}}
    bot.send_message(chat_id, "Оцените функцию распознавания достопримечательностей (1–5):")

@bot.message_handler(func=lambda msg: msg.chat.id in user_state)
def handle_rating(message):
    chat_id = message.chat.id
    text = message.text.strip()
    step = user_state[chat_id]['step']

    if not text.isdigit() or not 1 <= int(text) <= 5:
        bot.send_message(chat_id, "Пожалуйста, введите число от 1 до 5.")
        return

    rating = int(text)

    if step == 1:
        user_state[chat_id]['data']['recognition'] = rating
        bot.send_message(chat_id, "Оцените функцию построения маршрутов (1–5):")
        user_state[chat_id]['step'] += 1
    elif step == 2:
        user_state[chat_id]['data']['routes'] = rating
        bot.send_message(chat_id, "Оцените качество озвучки текста (1–5):")
        user_state[chat_id]['step'] += 1
    elif step == 3:
        user_state[chat_id]['data']['speech'] = rating
        bot.send_message(chat_id, "Оцените краткое описание достопримечательности (1–5):")
        user_state[chat_id]['step'] += 1
    elif step == 4:
        user_state[chat_id]['data']['summary'] = rating
        bot.send_message(chat_id, "Оцените общее впечатление о сервисе (1–5):")
        user_state[chat_id]['step'] += 1
    elif step == 5:
        user_state[chat_id]['data']['total'] = rating
        save_rating(chat_id, user_state[chat_id]['data'])
        bot.send_message(chat_id, "Спасибо за вашу оценку!")
        user_state.pop(chat_id)
```
